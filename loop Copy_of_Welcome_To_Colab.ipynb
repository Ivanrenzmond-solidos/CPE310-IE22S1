## **Loop Control**

# Break & Continue

After loop flow dynamically. 'Break' terminates the entire loop immediately. 'Continue' skips only the remaining code in the current iteration and jumps to the next cycle.

```
for i in range(1, 6):
if i == 3:
  continue
  print(i)
```

```
for i in range(1, 6):
    if i == 3:
        break
    print(i)
```
