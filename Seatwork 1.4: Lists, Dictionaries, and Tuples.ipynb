# 1. Create a list

```

fruit = ["apple", "banna", "cherry", "orange",]
print("Original list:", fruit)
```

# 2. Access elements

```

print("First fruit:", fruit[0])
print("Last fruit:", fruit[-1])
```

# 3. Add elements

```

fruit.append("mango")            # Add at the end
fruit.insert(1, "grape")         # Add at specific position
print("After adding fruits:", fruit)
```

# 4. Remove elements

```

fruit.remove("banna")            # Remove by value (corrected spelling)
popped = fruit.pop()             # Remove last item
print("After removing fruits:", fruit)
print("Popped item", popped)
```

# 5. Check if an item exists

```

if "apple" in fruit:
  print("Apple is in the list!")
```

# 6. Loop through the list

```
print("All fruits in the list:")
for item in fruit:
  print("-", item)
```

# 7. Length of the list

```
print("Total number of fruits:", len(fruit))
```

# 8. Sorting the list


```
fruit.sort()   # Alphabetical order
print("Sorted list:", fruit)
```

# 9. Reversing the list

```
fruit.reverse()
print("Reversed list:", fruit)
```
