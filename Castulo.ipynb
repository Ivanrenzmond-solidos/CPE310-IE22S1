## **Part 2 Conditional Execution**:

```python
# Exercise 1
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))

if hours > 40:
    regular_pay = 40 * rate
    overtime_pay = (hours - 40) * (rate * 1.5)
    pay = regular_pay + overtime_pay
else:
    pay = hours * rate

print("Pay:", pay)
```

```python
# Exercise 2
try:
    hours = float(input("Enter Hours: "))
    rate = float(input("Enter Rate: "))
    pay = hours * rate
    print("Pay:", pay)
except ValueError:
    print("Error, please enter numeric input")
```

```python
# Exercise 3: Score to Grade

try:
    score = float(input("Enter score: "))

    if score < 0.0 or score > 1.0:
        print("Bad score")
    elif score >= 0.9:
        print("A")
    elif score >= 0.8:
        print("B")
    elif score >= 0.7:
        print("C")
    elif score >= 0.6:
        print("D")
    else:
        print("F")

except ValueError:
    # Handles non-numeric input
    print("Bad score")
```
